```python
import pandas as pd
import spacy

# Load the spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample data loading, replace with your actual file
df = pd.read_csv("../data/pos_only_tagged_data.csv")

# Function to convert POS-tagged tuples back to plain text
def tuples_to_text(tuples_list):
    return ' '.join([word for word, tag in tuples_list])

# Convert each POS-tagged column to plain text
for column in df.columns:
    if "pos" in column.lower():  # Checking if the column contains POS-tagged data
        df[column.replace('_pos', '_text')] = df[column].apply(lambda x: tuples_to_text(eval(x)))

# Function to extract entities using spaCy
def extract_entities(text):
    doc = nlp(text)
    return [(ent.text, ent.label_) for ent in doc.ents]

# Apply NER to the reconstructed text columns and immediately print NER results
for column in df.columns:
    if "_text" in column:
        entity_column = column.replace('_text', '_entities')
        df[entity_column] = df[column].apply(extract_entities)
        # Displaying NER results for the current column
        print(f"NER results for {column}:")
        print(df[entity_column].head())  # Adjust the number of displayed rows as needed

# Optionally, save the DataFrame with extracted entities
df.to_csv("../data/enriched_dataset_with_entities.csv", index=False)
```
